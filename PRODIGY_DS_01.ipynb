```python
# Selecting columns from '1960' to '2022' inclusive
df_year = df.loc[:, '1960':'2022']

# Summing up values across all years
Total_year = df_year.sum()
```

```python
# Creating a figure with specified size
plt.figure(figsize=(20, 10))

# Creating a bar plot
plt.bar(Total_year.index, Total_year.values)

# Adding a title to the plot
plt.title('Total Values from 1960 to 2022')

# Adding labels to the x-axis and y-axis
plt.xlabel('Year')
plt.ylabel('Total Value')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Displaying the plot
plt.show()
```

```python
# Iterating over each column (year) in df_year
for i in df_year:
    # Printing the maximum population of the year
    print(f'The maximum population of year {i} is {df_year[i].max()}')
    # Printing the minimum population of the year
    print(f'The minimum population of year {i} is {df_year[i].min()}.\n')
```

```python
# Dropping missing values from the 'IncomeGroup' column and converting to string type
Income_group = data['IncomeGroup'].dropna().astype(str)

# Creating a histogram plot
plt.figure(figsize=(12, 6))
plt.hist(Income_group, bins=4, color='green', edgecolor='black', linewidth=1.2)

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Adding title to the plot
plt.title('Categories based on income')

# Displaying the plot
plt.show()
```

```python
# Dropping missing values from the 'Region' column and converting to string type
Regions = data['Region'].dropna().astype(str)

# Creating a histogram plot
plt.figure(figsize=(12, 6))
plt.hist(Regions, bins=7, edgecolor='black', linewidth=1.2)

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Adding title to the plot
plt.title('Categories based on regions')

# Displaying the plot
plt.show()
```
